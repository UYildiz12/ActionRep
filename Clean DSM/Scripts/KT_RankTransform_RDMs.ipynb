```python
from statsmodels.sandbox.stats.multicomp import multipletests

rdms = [rdm_sub1, LOTC_sub1, PPC_sub1, PMC_sub1]
rdm_names = ['rdm_sub1', 'LOTC_sub1', 'PPC_sub1', 'PMC_sub1']

# List of models
models = [opticflow, background, hmax, noofpeople, actionclass]
model_names = ['Opticflow', 'Background', 'Hmax_C1', 'Noofpeople', 'Actionclass']

df = pd.DataFrame(columns=['Comparison', 'Kendall\'s Tau', 'p-value'])
def get_upper_triangular(matrix):
    return squareform(matrix, checks=False)

for rdm, rdm_name in zip(rdms, rdm_names):
    if rdm_name!='rdm_sub1':
        tau, p_value = kendalltau(get_upper_triangular(rdm_sub1), get_upper_triangular(rdm))
        df = df.append({
            'Comparison': f'rdm_sub1 vs {rdm_name}',
            'Kendall\'s Tau': tau,
            'p-value': p_value
        }, ignore_index=True)
    for model, model_name in zip(models, model_names):
        tau, p_value = kendalltau(get_upper_triangular(rdm), get_upper_triangular(model))
        df = df.append({
            'Comparison': f'{rdm_name} vs {model_name}',
            'Kendall\'s Tau': tau,
            'p-value': p_value
        }, ignore_index=True)

# Apply Bonferroni correction
_, pvals_corrected, _, _ = multipletests(df['p-value'], method='bonferroni')
df['p-value-corrected'] = pvals_corrected

df = df.sort_values('Kendall\'s Tau', ascending=False)

fig = go.Figure(data=[
    go.Bar(
        name='Kendall\'s Tau',
        x=df['Comparison'],
        y=df['Kendall\'s Tau'],
        hovertext=df['p-value-corrected'].apply(lambda p: f"p-value: {'<0.001' if p < 0.001 else ('<0.01' if p < 0.01 else f'{p:.2f}')}"),
        marker_color=df['p-value-corrected'].apply(lambda p: 'black' if p < 0.01 else 'white')
    )
])

# Change the bar mode
fig.update_layout(barmode='group', title_text='Kendall\'s Tau for each comparison in Subject 1 (Bonferroni Corrected)')

fig.show()

# Or using double quotes to define the string
fig.write_image(r"C:\Users\UmurY\OneDrive\Masaüstü\Action_Clusters_Python\Result_Files\Clean DSM\Figures\KT_Sub1_WB_(r=0.5).png")
```
