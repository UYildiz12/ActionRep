```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

bhv_1=pd.read_csv(r'C:\Program Files (x86)\Umram\ActionClusters Project\ActionClusters(mixed)\Experiment_Folder\Output\Umur\bhvratings102_1.csv')
pairs_1=pd.read_csv(r'C:\Program Files (x86)\Umram\ActionClusters Project\ActionClusters(mixed)\Experiment_Folder\Output\Umur\sespairs102_1.csv')
bhv_2=pd.read_csv(r'C:\Program Files (x86)\Umram\ActionClusters Project\ActionClusters(mixed)\Experiment_Folder\Output\Umur\bhvratings102_2.csv')
pairs_2=pd.read_csv(r'C:\Program Files (x86)\Umram\ActionClusters Project\ActionClusters(mixed)\Experiment_Folder\Output\Umur\sespairs102_2.csv')

bhv_3=pd.read_csv(r'C:\Program Files (x86)\Umram\ActionClusters Project\ActionClusters(mixed)\Experiment_Folder\Output\Umur\bhvratings102_3.csv')

pairs_3=pd.read_csv(r'C:\Program Files (x86)\Umram\ActionClusters Project\ActionClusters(mixed)\Experiment_Folder\Output\Umur\sespairs102_3.csv')

merged_ratings = pd.concat([bhv_1, bhv_2, bhv_3], axis=0, ignore_index=True)
merged_pairs = pd.concat([pairs_1, pairs_2, pairs_3], axis=0, ignore_index=True)

unique_values = merged_pairs['mov1'].unique()
print(unique_values)


merged_pairs['mov2'] = merged_pairs['mov2'].str.replace(r'^.*\\Videos', '')
merged_pairs['mov2'] = merged_pairs['mov2'].str.replace('.avi', '')
merged_pairs['mov2'] = merged_pairs['mov2'].str.replace('\\', '')
merged_pairs['mov1'] = merged_pairs['mov1'].str.replace(r'^.*\\Videos', '')
merged_pairs['mov1'] = merged_pairs['mov1'].str.replace('.avi', '')
merged_pairs['mov1'] = merged_pairs['mov1'].str.replace('\\', '')


merged_pairs
merged_pairs[((merged_pairs['mov1'].str.contains('017')) & (merged_pairs['mov2'].str.contains('075'))) |
                           ((merged_pairs['mov1'].str.contains('075')) & (merged_pairs['mov2'].str.contains('017')))]



#Pairs of Null responses
list_null = merged_ratings[merged_ratings['response'].isnull()]
merged_pairs[merged_pairs.index.isin(list_null.index)]


merged_ratings['response'].fillna('3#', inplace=True)
merged_ratings[merged_ratings['response'].isnull()]

print(merged_pairs['mov1'].unique())
print(merged_pairs['mov2'].unique())


merged_pairs['mov1'] = merged_pairs['mov1'].str.replace(r'[^0-9]', '', regex=True)
merged_pairs['mov2'] = merged_pairs['mov2'].str.replace(r'[^0-9]', '', regex=True)


merged_ratings['response'].unique()
merged_ratings[merged_ratings['response']=='space']='3#'


merged_ratings['response'] = merged_ratings['response'].str.replace(r'[^0-9]', '', regex=True).astype(int)
print(merged_ratings)

merged_pairs['responses']=merged_ratings['response']





comm_vocal = [2, 28, 85, 95]
passive_inter = [4, 35, 58, 63, 89, 94]
manipulation = [7, 12, 13, 31, 50, 53, 96]
aim_in_space = [8, 20, 49, 71, 78, 87, 100]
skin_displacement = [15, 16, 23, 70]
digestion = [17, 21, 41, 72,88,92]
modify_substrate = [22, 44, 46, 64, 99]
active_inter = [36, 42, 47, 59, 75, 98]
comm_symbolic = [40, 51, 57]
locomotion_b = [55, 62, 68]
locomotion_q = [5,24,27,48,74, 77, 82]
list_order = [
    passive_inter, active_inter, comm_symbolic, comm_vocal,
    manipulation, modify_substrate, digestion,skin_displacement,
    aim_in_space, locomotion_b, locomotion_q
]
combined_lists = [num for sublist in list_order for num in sublist]
n = len(combined_lists)
df = pd.DataFrame(index=[f'A{num}' for num in combined_lists], columns=[f'A{num}' for num in combined_lists])
df.values[[range(n)], [range(n)]] = 7
print(df)




merged_pairs['mov1'] = merged_pairs['mov1'].str.replace(r'^0+', 'A', regex=True)
merged_pairs['mov2'] = merged_pairs['mov2'].str.replace(r'^0+', 'A', regex=True)
merged_pairs['mov1'] = merged_pairs['mov1'].str.replace(r'^0+', 'A', regex=True).replace('100', 'A100')
merged_pairs['mov2'] = merged_pairs['mov2'].str.replace(r'^0+', 'A', regex=True).replace('100', 'A100')


for row_idx, row in df.iterrows():
    for col_idx, value in row.iteritems():
        if pd.isnull(value):
            response_row = merged_pairs[((merged_pairs['mov1'] == row_idx) & (merged_pairs['mov2'] == col_idx)) | ((merged_pairs['mov1'] == col_idx) & (merged_pairs['mov2'] == row_idx))]
            if not response_row.empty:
                df.at[row_idx, col_idx] = response_row['responses'].values[0]


merged_ratings[(merged_pairs['mov1'] == 'A68') & (merged_pairs['mov2'] == 'A13')]

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns, index=df.index)

plt.figure(figsize=(14, 12))
heatmap=sns.heatmap(df_scaled, cmap='coolwarm_r')
plt.show()


df = df.apply(pd.to_numeric, errors='coerce')


```

```python
import matplotlib.patches as mpatches

plt.figure(figsize=(16, 14))
heatmap=sns.heatmap(df_scaled, cmap='coolwarm_r')
yticks = heatmap.get_yticklabels()
xticks = heatmap.get_xticklabels()
groups = [
    ('passive_inter', [4, 35, 58, 63, 89, 94]),
    ('active_inter', [36, 42, 47, 59, 75, 98]),
    ('comm_symbolic', [40, 51, 57]),
    ('comm_vocal', [2, 28, 85, 95]),
    ('manipulation', [7, 12, 13, 31, 50, 53, 96]),
    ('modify_substrate', [22, 44, 46, 64, 99]),
    ('digestion', [17, 21, 41, 72,88,92]),
    ('skin_displacement', [15, 16, 23, 70]),
    ('aim_in_space', [8, 20, 49, 71, 78, 87, 100]),
    ('locomotion_b', [55, 62, 68]),
    ('locomotion_q', [5,24,27,48,74, 77, 82])
]

colors = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'olive', 'cyan', 'magenta']

start = 0
for j,(group_name,indexes) in enumerate(groups):
    end = start + len(indexes)
    for i in range(start,end):
        yticks[i].set_color(colors[j])
        xticks[i].set_color(colors[j])
    start = end


heatmap.set_yticklabels(yticks)
patches = [mpatches.Patch(color=colors[i], label=name) for i,(name,indexes) in enumerate(groups)]
legend = plt.legend(handles=patches,bbox_to_anchor=(1.20,1), loc=2)
plt.setp(legend.get_texts(), fontsize='x-large') 

plt.show()
```
